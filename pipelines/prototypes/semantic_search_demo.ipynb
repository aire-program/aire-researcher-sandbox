# Semantic Search Demo

Demonstrates semantic search capabilities.

```python
import sys
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

if 'google.colab' in sys.modules:
    !pip install -q pandas scikit-learn
    !git clone https://github.com/YOUR_USERNAME/aire-researcher-sandbox.git
    %cd aire-researcher-sandbox

df = pd.read_csv('data/sample_texts/articles_sample.csv')
tfidf = TfidfVectorizer(stop_words='english')
matrix = tfidf.fit_transform(df['abstract'].fillna(''))
query_vec = tfidf.transform(["climate change"])
sims = cosine_similarity(query_vec, matrix)
print("Top match score:", sims.max())
```
